```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib as mpl
from urllib.request import urlopen
```

```python
# Define the MLB Top 100 Prospects URL
url = "https://www.mlb.com/prospects/top100/"
```

```python
from urllib.request import Request, urlopen

req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
webpage = urlopen(req).read()
```

```python
# Print the HTML code
from IPython.core.display import html

# Extract html
html = webpage.decode("utf-8")


print((html))
```

title_index = html.find("<title>")

```python
from bs4 import BeautifulSoup
import requests

    
# Parse the html content
soup = BeautifulSoup(html,  "html.parser")
 
# Find all li tag
prospects = soup.find_all("</div")
 
# Iterate through all li tags
for prospect in prospects:
    # Get text from each tag
    print(prospect.text)
 
print(f"Total {len(prospects)} prospect tags found")
```
